### ***Variables categricas debajo del modelo***

---

Debe notar que hay una docena de coeficientes para `District` arriba. Esto se debe a que es una variable categrica y, para las variables categricas, se obtiene un coeficiente para todas las categoras menos una. Si solo hay dos categoras (por ejemplo, gnero), la caracterstica se convierte en una columna de ceros y unos antes de introducirla en la regresin, donde se trata como una variable numrica regular.

Cuando hay ms de dos categoras, una categora se designa como la categora de "referencia" o "lnea de base", y se crean columnas `dummy` de unos y ceros para todas las dems categoras. Tomemos un ejemplo ficticio con tres categoras y cinco registros:

| Categorica   |
|--------------|
| A            |
| B            |
| C            |
| C            |
| A            |

Antes de ajustar la regresin lineal, la columna `Categorica` se transforma en **dos** columnas `dummys` (no se agrega una columna para la categora de referencia).

La primera columna es 1 si es `B` y 0 en caso contrario, mientras que la segunda columna es 1 si el distrito es `C` y 0 en caso contrario. Obtenemos:

| Categorica_B | Categorica_C |
|--------------|------------- |
| 0            | 0            |
| 1            | 0            |
| 0            | 1            |
| 0            | 1            |
| 0            | 0            |

Las columnas `dummys` se introducen en la regresin lineal y se tratan como variables numricas regulares. Esta tcnica se llama **one-hot encoding** y se puede hacer manualmente con el mtodo de `pandas`, `pd.get_dummies()`:




````python
## Ejemplo dummyzar una variable categrica.
df_dummy=pd.DataFrame({'Category':['A','B','C','C','A']})
pd.get_dummies(df_dummy,columns=['Category'], drop_first=True) # drop_first = True, elimina la
                                                               # la primera categora
````

## ***Evaluacin del rendimiento predictivo***

---

Para cualquier conjunto de datos dado, el residuo es la diferencia entre la prediccin del modelo para ese registro y el valor real de ese punto. El **error absoluto medio (MAE)** es una mtrica que resume el rendimiento holstico (del todo) del modelo en toda la data. El MAE se calcula tomando el valor absoluto de cada residuo y luego tomando la media de todos esos valores absolutos. En esencia, el MAE describe la magnitud tpica de los residuales; cuanto menor sea el MAE, mejor.

````python
## Funcin que calcula el MAE (Mean Absolute Error) de un modelo.
def MAE(true_values, prediction):
    '''
    Input: true_values : Valores de la varible respuesta en el conjunto X.
           prediction : Valores ajustados del modelo en el conjunto X.
    Output: MAE (Error absoluto medio) del modelo sobre el conjunto X.
    '''
    return np.mean(np.abs(true_values-prediction))
##
MAE_model_log_train = MAE(model_log.predict(X_train), np.log(X_train['Sale_price']))
MAE_model_log_test = MAE(model_log.predict(X_test), np.log(X_test['Sale_price']))
##
MAE_exp_modelo_log_train = MAE(np.exp(model_log.predict(X_train)) ,X_train['Sale_price'])
MAE_exp_modelo_log_test = MAE(np.exp(model_log.predict(X_test)) ,X_test['Sale_price'])
##
print('MAE model_log y log de los verdaderos valores de Sale price en TRAIN:',MAE_model_log_train)
print('MAE model_log y log de los verdaderos valores de Sale price en TEST: ',MAE_model_log_test)
print('-'*90)
print('MAE exp(model_log) y los verdaderos valores de Sale price en TRAIN:',MAE_exp_modelo_log_train)
print('MAE exp(model_log) y los verdaderos valores de Sale price en TEST:',MAE_exp_modelo_log_test)
````

Otra opcin muy utilizada para evaluar la regresin es  **la raz del error cuadrtico medio (RMSE)**. Esto se calcula tomando el cuadrado de cada residuo, promediando y luego sacando la raz cuadrada. A continuacin se muestra la implementacin en python, que posiblemente sea ms fcil de entender que la explicacin anterior:

````python
## Funcin que calcula el RMSE (Root Mean Squared Error) de un modelo.
def RMSE(true_values,prediction):
    '''
    Input: true_values : Valores de la varible respuesta en el conjunto X.
           prediction : Valores ajustados del modelo en el conjunto X.
    Output: RMSE (raz del error medio cuadrtico) del modelo sobre el conjunto X.
    '''
    return np.sqrt(np.mean(np.square(true_values-prediction)))
##
RMSE_exp_model_log_train = RMSE(X_train['Sale_price'], np.exp(model_log.predict(X_train)))
RMSE_exp_model_log_test = RMSE(X_test['Sale_price'], np.exp(model_log.predict(X_test)))
##
print('RMSE exp(model_log) y Sale_price en TRAIN:',RMSE_exp_model_log_train)
print('RMSE exp(model_log) y Sale_price en TEST:',RMSE_exp_model_log_test)
````

Finalmente, est el **error porcentual absoluto medio**. Esto toma el valor absoluto de cada residual y lo divide por el valor real de ese punto para obtener un porcentaje, luego promediando todos los porcentajes.



````python
## Funcin que calcula el MAPE (Mean Adsolute Error Percentage) de un modelo.
def MAPE(true_value, prediction):
    return np.mean(np.abs((true_value-prediction)/true_value)*100)
##
MAPE_exp_modelo_log_train = MAPE(X_train['Sale_price'], np.exp(model_log.predict(X_train)))
MAPE_exp_modelo_log_test = MAPE(X_test['Sale_price'], np.exp(model_log.predict(X_test)))
##
print('MAPE entre model_log y log de los verdaderos valores de Sale_price:', MAPE_exp_modelo_log_train)
print('-'*90)
print('MAPE entre exp(model_log) y los verdaderos valores de Sale_price:  ', MAPE_exp_modelo_log_test)
````

### ***Pregunta:***

---

Por qu no usar simplemente MAE en lugar de tener que lidiar con tres mtricas diferentes?



Aunque `RMSE` puede parecer innecesariamente complicado en comparacin con MAE, se usa mucho ms comnmente. La razn es que la mtrica `RMSE` es la misma que se minimiza en el set de train mediante un modelo de regresin lineal estndar (tambin llamado regresin de **mnimos cuadrados ordinarios (OLS)**). Esto lo convierte en una opcin "natural" que utiliza la misma mtrica para evaluar el rendimiento fuera de la muestra (`test`). Considere lo siguiente:

1. El `RMSE` pone mucho ms peso en los valores atpicos, ya que los errores se elevan al cuadrado antes de promediarlos. En los casos en que los valores atpicos son especialmente malos y deben castigarse, el `RMSE` es una mejor opcin.

2. El `MAPE` tiene una buena interpretacin en la que podemos decir que las predicciones de un modelo son, en promedio, incorrectas en un cierto porcentaje. Por ejemplo, nuestro modelo de precio de la vivienda podra tener un promedio de descuento del 30%. Dicho modelo a menudo se describe como "70% preciso".

Las siguientes ecuaciones definen las tres mtricas de error anteriores:

\begin{eqnarray*}
    MAE &= \frac{1}{n} \sum_{i=1}^n \left| y_i - \hat{y}_i \right| \\
    RMSE &= \sqrt{ \frac{1}{n} \sum_{i=1}^n ( y_i - \hat{y}_i )^2 } \\
    MAPE &= 100\% \frac{1}{n} \sum_{i=1}^n \left|\frac{ y_i - \hat{y}_i}{y_i} \right|
\end{eqnarray*}

````python
## con sklearn.metrics
## 1.MAE
mae_train = mean_absolute_error(np.exp(model_log.predict(X_train)),X_train['Sale_price'],)
mae_test = mean_absolute_error(np.exp(model_log.predict(X_test)),X_test['Sale_price'])
print('='*30)
print(f'MAE en TRAIN: {mae_train:.3f}')
print(f'MAE en TEST:  {mae_test:.3f}')
print('='*30)
## 2.RMSE
mse_train = mean_squared_error(np.exp(model_log.predict(X_train)),X_train['Sale_price'],squared=False)
mse_test = mean_squared_error(np.exp(model_log.predict(X_test)),X_test['Sale_price'], squared=False)
print(f'RMSE en TRAIN: {mse_train:.3f}')
print(f'RMSE en TEST:  {mse_test:.3f}')
print('='*30)
## 3. MAPE
mape_train = mean_absolute_percentage_error(X_train['Sale_price'], np.exp(model_log.predict(X_train)))
mape_test = mean_absolute_percentage_error(X_test['Sale_price'],np.exp(model_log.predict(X_test)))
print(f'MAPE en TRAIN: {mape_train*100:.3f}')
print(f'MAPE en TEST:  {mape_test*100:.3f}')
print('='*30)
````

Ilustremos las diferencias entre MAE, RMSE y MAPE con un ejemplo. En los siguientes datos, nuestra muestra `test` ha sido contaminada con un solo valor atpico. Comparar  `MAE`, `RMSE` y `MAPE` de los modelos:

````python
np.random.seed(135568109)
error = np.random.normal(0,0.1,100)   ## Simulacin de 100 valores de N(0,(0.1))
x_data = np.arange(0,10,0.1)          ## 100 valores equidistantes entre 0 y 10
y_data = x_data + error               ## 100 valores con una perturbacin.
##
train_ex = np.random.randint(0,99,80) ## 80 valores entre 0 y 98
x_train = x_data[train_ex]            ## Datos de train para la variable explicativa.
y_train = y_data[train_ex]            ## Datos de train para la variable respuesta.
##
test_ex = list(set(range(100))-set(train_ex))
x_test = x_data[test_ex]
y_test = y_data[test_ex]
##
x_test = np.append(x_test,11)          ## La contaminamos.
y_test = np.append(y_test,1000)        ## La contaminamos.
````

`MAE` y `MAPE` estn a favor del `model1`, mientras que `RMSE` est a favor del `model2`! Observe cmo un nico valor atpico puede suponer una amenaza para nuestros modelos. Si eliminamos el valor atpico claro del conjunto de datos, obtenemos:

````python
model1_pred = x_test[:-1]
model2_pred = (7.8182)*x_test[:-1]-24.3557

print('-----------------------------------')
print("MAE de model1:", MAE(model1_pred,y_test[:-1]))
print("MAE de model2:", MAE(model2_pred,y_test[:-1]))
print('-----------------------------------')
print("RMSE de model1:", RMSE(model1_pred,y_test[:-1]))
print("RMSE de model2:", RMSE(model2_pred,y_test[:-1]))
print('-----------------------------------')
print("MAPE de model1:", MAPE(model1_pred,y_test[:-1]))
print("MAPE de model2:", MAPE(model2_pred,y_test[:-1]))
print('-----------------------------------')
````

Podemos ver claramente que `model1` se ajusta mucho mejor a los datos. Observe cun robusto fue `MAPE` contra los valores atpicos: los valores no cambiaron mucho aqu.

## ***Colinealidad y errores estndar.***

---

El modelo de la seccin anterior parece ser un buen paso hacia nuestro objetivo principal.En las ltimas lecturas se aprendi que agregar una gran cantidad de nuevas varibales a un modelo y luego dejarlo as,no es una estrategia prudente, pues:

1. No queremos incluir caractersticas que agreguen poco poder predictivo a nuestro modelo. De lo contrario, podemos terminar con un modelo demasiado complicado. Recuerde que generalmente queremos un modelo que logre un buen nivel de predictividad sin sacrificar la simplicidad.

2. Agregar demasiadas caractersticas aumenta el peligro de *overfitting*, lo que disminuye la aplicabilidad potencial del modelo.

Hemos aprendido un par de mtodos para aliviar estos problemas:

1. Podemos utilizar indicadores como el `AIC` para detectar si estamos sumando variables con poco poder predictivo.

2. Podemos dividir nuestros datos de muestra en conjuntos `Train/Test` y usar el conjunto `Test`, junto con alguna medida como `RMSE`, para evaluar la aplicabilidad de nuestro modelo.

Sin embargo, existe un tercer problema potencial cada vez que agregamos demasiadas caractersticas a nuestros modelos:

3. La regresin lineal en realidad puede **desestabilizarse** cuando hay demasiados predictores que no agregan valor.

Esto significa que pequeos cambios en los datos de entrenamiento introducidos en el modelo dan como resultado grandes cambios en los coeficientes ajustados; el sobreajuste es una consecuencia comn de este fenmeno.

### ***Pregunta:***

Qu crees que hace que suceda en lo anterior?

A medida que se agregan ms predictores, comienzan a contribuir con informacin redundante (lo que explica por qu cada predictor adicional aporta informacin marginal cero). Matemticamente, esto significa que los nuevos predictores a menudo se pueden expresar en buena medida como combinaciones lineales de algunos de los otros predictores. Cuando esto sucede, se dice que un modelo exhibe **colinealidad** o **multicolinealidad**.

Dado que los predictores se vuelven en su mayora redundantes, esto significa que se pueden obtener predicciones casi idnticas con valores de coeficientes muy diferentes, mediante el "intercambio" entre los coeficientes de los predictores que exhiben colinealidad.
En consecuencia, el modelo no tiene forma de decidir solo a partir de los datos cul de estos posibles modelos con predicciones idnticas elegir, lo que se traduce en errores estndar inflados y un alto riesgo de ***overfitting***.

Para demostrar este efecto, introduciomos la varible.
La pequea constante est ah para evitar que la nueva funcin sea *perfectamente* colineal con log(`Fin_sqft`) y log(`Lotsize`) (lo que creara un error de "dividir por cero" en el cdigo de la funcin de regresin en segundo plano):




````python
# modelo base sin caracterstica colineal diseada
model_finlot = smf.ols(formula = 'np.log(Sale_price) ~ '
                           'Stories + Year_Built'
                           '+ np.log(Fin_sqft)'
                           '+ np.log(Lotsize)'
                           '+ C(Style)'
                           '+ Extwall  + Units + Bdrms'
                           '+ Fbath + Hbath + C(Nbhd)'
                           '+ C(Sale_year)',
                 data = X_train).fit()
print(model_finlot.summary())
````

````python
model_coll = smf.ols(formula = 'np.log(Sale_price) ~ '
                           'Stories + Year_Built'
                           '+ np.log(Fin_sqft)'
                           '+ np.log(Lotsize)'
                           '+ np.log(Fin_sqft/Lotsize + 0.0001)'
                           '+ Style'
                           '+ Extwall  + Units + Bdrms'
                           '+ Fbath + Hbath'
                           '+ C(Nbhd)'
                           '+ C(Sale_year)',
                 data = X_train).fit()
print(model_coll.summary())
````

Para detectar e investigar las colinealidades, es til observar las correlaciones entre los predictores dentro de un modelo, como se muestra a continuacin. Observe las altas correlaciones (positivas *o* negativas) entre la funcin de ingeniera `np.log(Fin_sqft / Lotsize + 0.0001)` y `np.log(Fin_sqft)`/`np.log(Lotsize)`. Tenga en cuenta tambin las otras correlaciones altas con baos y nmero de pisos en la matriz. Estos pueden desestabilizar el modelo y dificultar la formulacin de afirmaciones significativas sobre el valor incremental de agregar baos e historias adicionales.

````python
M = model_coll.model.data.exog # La matriz de diseo X  con
                               # predictores que incluyen variables "dummys"
                               # para las variables categricas.
param_names = model_coll.model.data.param_names # Nombre de los parmetros.
M_df = pd.DataFrame(M, columns=param_names) # Pandas DataFrame
# Subconjunto de predictores para graficar:
M_sub=M_df[['np.log(Fin_sqft)','np.log(Lotsize)',
              'np.log(Fin_sqft / Lotsize + 0.0001)',
              'Bdrms','Fbath','Hbath','Stories']]
plt.figure(figsize=(6,5))
sns.heatmap(M_sub.corr(), cmap="RdYlBu",annot=True, square=True, vmin=-0.8, vmax=0.8, fmt="+.1f")
plt.title("Correlaciones entre predictores")
plt.show()
````

Esto ilustra el peligro de simplemente incluir caractersticas adicionales en un modelo. Una forma de lidiar con esto es lo que discutimos en casos anteriores: usar el criterio AIC para evaluar iterativamente cada caracterstica nueva y determinar si realmente agrega suficiente valor incremental adems de una penalizacin por la complejidad adicional. Otra es examinar matrices de correlacin como la que creamos anteriormente para identificar cualquier valor cercano a -1 o 1. Con tcnicas como ***cross-validation*** y la ***regularizacin*** se pueden tener formas adicionales de lidiar con modelos complejos y potencialmente sobreajustados.
