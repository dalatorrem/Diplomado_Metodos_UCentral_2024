```python
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import pandas as pd
```

## GDP

US Quarterly GDP for the past 70 years. <br>
Producto interno bruto (Gross domestic product)

```python
GDP = pd.read_csv('https://github.com/dalatorrem/Diplomado_Metodos_UCentral_2023/raw/main/data/GDP.csv', parse_dates=['DATE'])
GDP.info()
```

```python
px.line(GDP, x='DATE', y='GDP')
```

```python
GDP['DATE'].dt.year
```

```python
GDP['YEAR'] = GDP['DATE'].dt.year
GDP['MONTH'] = GDP['DATE'].dt.month
GDP
```

```python
piv_gdp = pd.pivot_table(GDP, index='MONTH', columns='YEAR', values='GDP', aggfunc='mean')
piv_gdp
```

```python
px.imshow(piv_gdp)
```

## Influenza

```python
ILI = pd.read_csv('https://github.com/dalatorrem/Diplomado_Metodos_UCentral_2023/raw/main/data/CDC.csv')
ILI['date'] = ILI['Year']+ILI['Week']/52
ILI.head()
```

```python
px.line(ILI, x='date', y=['Percent of Deaths Due to Pneumonia and Influenza',
            'Expected',
            'Threshold'])
```

```python
piv_ili = pd.pivot_table(ILI, columns='Week', index='Year',
                         values='Percent of Deaths Due to Pneumonia and Influenza', aggfunc='mean')
piv_ili
```

```python
px.imshow(piv_ili)
```

## Airline Passengers

The number of passenters in international flights in the 50s displays both a trend and a seasonal pattern

```python
airline = pd.read_csv('https://github.com/dalatorrem/Diplomado_Metodos_UCentral_2023/raw/main/data/international-airline-passengers.csv', sep=';')
airline['DATE'] = pd.to_datetime(airline['Month'])
airline['YEAR'] = airline['DATE'].dt.year
airline['Month'] = airline['DATE'].dt.month
airline.head()
```

```python
px.line(airline, x='DATE', y='Passengers')
```

```python
piv_air = pd.pivot_table(airline, columns='Month', index='YEAR',
                         values='Passengers', aggfunc='mean')
piv_air
```

```python
px.imshow(piv_air)
```
