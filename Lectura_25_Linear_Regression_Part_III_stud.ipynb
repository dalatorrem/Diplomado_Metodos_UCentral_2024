### ***Inspeccin de los residuales para diagnosticar un modelo ajustado.***

La regresin lineal es conceptualmente simple, ajustamos una superficie a nuestros datos an no todo es tan simple como eso. ***El anlisis de residuales*** es una herramienta importante para diagnosticar muchos problemas que pueden afectar una regresin lineal.
Los residuales son la diferencia entre las observaciones $y$ y los valores ajustados correspondientes $\hat{y}$; visualmente es la distancia entre la superficie ajustada y los datos.

El anlisis residual nos permite probar algunos de los supuestos tericos que sustentan la regresin lineal. La versin corta es que la regresin lineal funciona mejor cuando los residuos estn [distribuidos normalmente](https://en.wikipedia.org/wiki/Normal_distribution).

Hemos mencionado que "la regresin lineal supone que los datos se distribuyen normalmente", lo cual no es del todo correcto. La regresin lineal an puede ser una herramienta til y poderosa (y tericamente justificada!) incluso si los datos se desvan de esta suposicin. Dicho esto, una distribucin con colas anchas es un problema particular para la regresin lineal, porque los puntos "en las colas" que estn lejos de sus valores ajustados pueden afectar de manera desproporcionada los coeficientes ajustados y las predicciones. por ejemplo, ser causado por un punto decimal fuera de lugar cuando un empleado distrado ingres manualmente los datos sin procesar en una hoja de clculo, puede causar estragos en un modelo Se dice que dicho punto de datos tiene una alto **leverage** en el modelo.

Otra situacin problemtica que se puede detectar mediante el anlisis de residuos es la **heteroscedasticidad**, lo que significa que los residuos tienen una variacin pequea en algunos subconjuntos de datos y una variacin alta en otros. Vimos esto exactamente desde el principio en el caso anterior, con los puntos de datos "desplegndose" alrededor de la lnea de mejor ajuste, y determinamos que esto causara problemas. Lo opuesto a la heterocedasticidad es la **homocedasticidad**, que es lo que queremos ver en los datos y significa que los residuos tienen una varianza similar en todos los subconjuntos de datos.

### ***Transformando variables para mitigar problemas de residuos.***

Ya sabemos que los residuos son problemticos en este caso, que hacemos?. Una posible solucin es realizar una transformacin de la variable de respuesta. Intentemos hacer que la variable de respuesta sea el logaritmo de `Sale_price`. Es una suposicin natural que el precio de venta debe ser proporcional a los pies cuadrados de la propiedad. Por lo tanto, tambin tomaremos el logaritmo de los pies cuadrados como variable predictora en este modelo:

````python
model_log = smf.ols(formula = "np.log(Sale_price) ~ C(District) + Units+ np.log(Fin_sqft)",
                 data = data_clean).fit()
print(model_log.summary())
````

### ***Tratamiento de valores atpicos.***

Nuestro anlisis de residuos muestra una serie de valores atpicos. Como se mencion anteriormente, los valores atpicos bajos no son del todo inesperados; p.ej. cuando las personas venden una propiedad a amigos o familiares, o si la propiedad est en mal estado, el precio de venta podra ser significativamente ms bajo de lo esperado por el modelo. No obstante, los valores atpicos pueden desestabilizar un modelo y reducir significativamente su capacidad predictiva a pesar de que solo representan un subconjunto marginal de los datos generales.

### ***Pregunta:***

Qu debe hacer un analista con los valores atpicos?

---
Es tentador simplemente eliminar puntos de datos con grandes residuos; sin embargo, este punto de vista adolece de un sesgo retrospectivo. Los valores atpicos con residuos altos no se pueden determinar antes de que se cree el modelo, lo que significa que no se pueden eliminar por adelantado.

Ms bien, el punto de partida siempre debe ser inspeccionar los datos sin procesar iniciales en busca de puntos de datos con valores inusualmente pequeos o grandes para ciertas caractersticas y comprender por qu son diferentes. Si hay algo claramente errneo con esos datos (como un punto decimal fuera de lugar), o si el motivo de estos valores pequeos o grandes puede explicarse efectivamente a travs de un factor externo que no es capturado por los datos en s, entonces esto justifica eliminarlos. De lo contrario, es mejor tener en cuenta estos puntos de datos durante todo el proceso de modelado y tratarlos durante el proceso de modelado en s.
