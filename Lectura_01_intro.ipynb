## Introducción
---

Se tienen avances gigantescos en el poder de la computación y la interconectividad de dispositivos durante las últimas décadas. Esto, nos a llevado al punto en que podemos aprovechar el poder de una gran cantidad de datos para tomar decisiones mejores y más rápido. El campo de trabajo dedicado a hacer todo lo necesario para que esto suceda se ha convertido en lo que ahora se conoce como **Ciencia de datos**.

La ciencia de datos está en auge en este momento, con muchas oportunidades de innovación y colocación laboral en todas las industrias. Pero, ¿quién es exactamente un **Científico de datos**? ¿Qué herramientas utilizan? ¿Cuál es su labor?

---
## Ciencia de datos: ¿Es el mejor nombre?
---

_**La gran mayoría de los trabajos que aprovechan las habilidades de manejo de datos NO son trabajos de científico de datos**_. Esto significa que el término **Ciencia de datos** es un nombre inapropiado. Muchas oportunidades laborales como : "Analista de datos", "Ingeniero de datos" y "Analista de inteligencia empresarial" (entre muchos más) tienen un orden de magnitud más de puestos vacantes que **Científico de datos** y, al mismo tiempo, aprovechan las habilidades de datos en gran medida. De hecho, una gran cantidad de trabajos de **Científico de datos** están mal clasificados y encajarían mejor en una de las otras tres familias de trabajos mencionadas anteriormente.

Un **Profesionales de datos** será quien combina conocimientos en:

*   **Matemáticas y estadística.** Las habilidades de análisis y modelado, que es el arte de inferir, pronosticar, deducir y representar cosas sobre el mundo utilizando principios lógicos. En este caso, a veces es necesario o muy útil conocimientos matemáticos variados.
*   **Ciencias de la Computación.** A partir de aquí, necesitamos la habilidad de codificar o programar, que es la práctica de escribir instrucciones que una computadora que puede entender para dar vida a nuestro análisis y modelado.

*  **Experticia empresarial.** Necesario guiar nuestro análisis y/o modelado para la toma de decisiones dados los resutados.

## El lenguaje de programación
---

Para lograr que una computadora realice la tarea que tenemos en mente, debemos comunicarnos con ella, buscando un punto común. Este lenguaje común se conoce como **lenguaje de programación**. Un lenguaje de programación traduce las instrucciones legibles por un ser humano en instrucciones legibles por el computador, es decir el **código de máquina**.

Se debe resaltar que en general los **lenguajes de programación** son mucho más estrictos en la aplicación de estas reglas que los lenguajes entre seres humanos. Si no se siguen exactamente las reglas de un lenguaje de programación, su programa no **compilará**, es decir, el computador no logró traducir sus instrucciones a **código de máquina**. Pero esto hace que un gran grupo de personas pueda leer y trabajar en el código.

Hay muchos, muchos lenguajes de programación, pero para este curso y para gran parte del trabajo de datos, hay un rey por encima de todo ...

## ¿Por qué Python?
---
El lenguaje de programación de elección para el trabajo con datos hoy en día es **Python**. (Hay otros como R, Julia, C, C++).

**Python** es muy adecuado pues: 
 
1.   Es fácil de aprender.
2.   Es fácil de leer (¡otros lenguajes de programación no!).
3.   Es **código abierto**, lo que significa que es de uso gratuito. Muy importante: muchos en la comunidad de **Python** adicionan constantemente más funciones al lenguaje sin ningún costo.

El item 3. determina que quizá, lo que se necesita hacer ya está realizado y compartido por alguien más. Muchas de las funcionalidades ya está listas para ser utilizadas y no necesariamente se deben programar. Podemos encontrar por ejemplo en una tabla de datos o `dataframe` el valor promedio o media, así como el máximo o mínimo, o quizá algo útil que desee hacer ya alguien lo programó y puede ser utilizado.

Estas funcionalidades escritas por otras personas y utilizables se denominan **librerias** o **paquetes**. **Python** tiene librerias para casi cualquier cosa que se nos venga en mente. Del ejemplo anterior, podemos usar la librería `numpy` para cálculos matemáticos. Para interactuar con archivos Excel se puede usar `openpyxl`. Se estima que hay más de 200.000 **librerias** de **Python** con libre acceso.

## El IDE
---

Es posible que haya notado que el archivo que está leyendo actualmente vive dentro de un "contenedor" llamado **Jupyter** o **Collaboratory de google**.

Para comprender esto, primero debemos comprender qué es un **Entorno de desarrollo integrado (IDE: Integrated Development Environment)**. Un IDE es la herramienta principal que buscará al escribir código. Los archivos de código en realidad no son más que archivos de texto y también podrían escribirse en el Bloc de notas o cualquier otro editor de texto . Pero ninguno de esos editores de texto básicos proporciona otras funcionalidades de codificación esenciales, como verificar si escribió una palabra clave donde ella debe estar o ejecutar su código para probar si funciona.

Esto significa que los IDE son esencialmente editores de texto con super poderes, con una funcionalidad que ayuda enormemente al proceso de codificación. Uno de los más populares para **Python** y el trabajo de datos es **Jupyter** porque hace algo que casi ningún otro IDE permite: le permite escribir y ejecutar código junto con texto estándar.
