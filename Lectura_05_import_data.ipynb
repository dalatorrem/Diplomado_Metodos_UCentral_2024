## Máscaras
---
Supongamos que deseamos obtener los regístros que cumplen cierta condición. Por ejemplo, todos aquellos donde el género sea mujer, las personas menores de 30 años, aquellos que ganaron plata, o quizá mujeres que ganaron plata y son menores de 30 años.

Para hacer esto necesitamos una serie de `pandas` de tipo booleano, que determina que registros cumplen la condición deseada.

```python
df.head()
```

```python
mask = (df['Year'] > 1950)
mask
```

```python
## Tenemos todos los registros de quienes ganaron medalla después del Año 1948
df[mask]
```

```python
mask_1 = (df['Edad_Atleta'] <= 20)
```

```python
df_menores_de_20 = df[mask_1]
display(df_menores_de_20.describe().T)
df_menores_de_20.head()
```

```python
df_menores_de_20.reset_index(drop=True, inplace=True)
df_menores_de_20
```

```python
## Podemos juntar ambas condiciones
df[mask & mask_1]#.describe().T
```

```python
df['Medalla'].value_counts()
```

```python
## informació para un segmento particular.
df[mask & mask_1]['Medalla'].value_counts()
```

```python
mask_2 = df['Country'] == 'China'
df[mask_1 & mask_1 & mask_2].reset_index(drop=True)['Género'].value_counts(normalize=True)*100
```

## `pandas.crosstab`
---
La función `pandas.crosstab` devuelve la tabla de contingencia resultante de cruzar dos o más columnas de un dataframe. Por defecto ell resultado evalúa las frecuencias (absolutas o relativas) de cada combinación de valores, es posible especificar una función de agregación.

```python
pd.crosstab(df['Género'],df['Medalla'])
```

```python
## Podemos obtener la suma por filas y columnas (subtotales) y renombrar esta
pd.crosstab(df['Género'],df['Medalla'], margins=True, margins_name='Total')
```

```python
## Podemos mostrar las frecuencias 'normalizadas' (es decir una parte de la unidad). Se puede obtener respecto al total o con respecto a los subtotales
## de filas o columnas. Asignamos al parámetro normalize el valor True, la normalización se realiza con respecto al total de muestras
pd.crosstab(df['Género'],df['Medalla'], normalize=True)
```

Del total de las medallas entregadas: 

1.   19.8% fueron de bronce y entregas a hombres.
2.   12.2% fueron de plata y entregadas a mujeres.

En cada casilla podemos extraer información similar.


```python
pd.crosstab(df['Género'],df['Medalla'], normalize=True, margins=True, margins_name= 'Total %')
```

```python
pd.crosstab(df['Género'],df['Country'])
```

```python
pais_genero = pd.DataFrame(pd.crosstab(df['Género'],df['Country']).unstack()).rename({0:'Conteo'},axis=1)
pais_genero.head(10)
```

### ***Tablas de Wikipedia***

```python
import requests
from bs4 import BeautifulSoup

url = 'https://es.wikipedia.org/wiki/Chicago_Bulls'
r = requests.get(url)
print(f'Response get --> {r}')
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

### ***Tabla with BeautifulSoup.***

```python
print(type(soup.find_all('table')))
tables = [x for x in soup.find_all('table')]
```

```python
lista = []
for i in tables[0].find_all('tr'):
    try:
        i.find('td').get_text()
    except:
        lista.append(np.nan)
    else:
        lista.append(i.find('td').get_text())
#lista, len(lista)
```

```python
lista1 = []
for i in tables[0].find_all('tr'):
    try:
        i.find('th').get_text()
    except:
        lista1.append(np.nan)
    else:
        lista1.append(i.find('th').get_text())
#lista1, len(lista1)
```

```python
dicc = dict(zip(lista1, lista))
pd.DataFrame(dicc.items(), columns=['categoria', 'info'])
```

```python
result = pd.DataFrame(list(zip(lista1, lista)), columns=['col1', 'col2'])
result['col2'] = result['col2'].apply(lambda x: str(x).replace('\n',""))
result = result.dropna(subset=['col1'])
result
```
