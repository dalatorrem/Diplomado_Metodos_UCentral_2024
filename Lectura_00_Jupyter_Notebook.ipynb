## ¿Qué son los Jupyter Notebooks?
---

Jupyter es un entorno de desarrollo interactivo basado en la web que admite varios lenguajes de programación, sin embargo, se usa más comúnmente con el lenguaje de programación Python.

El entorno interactivo que proporciona Jupyter permite a los estudiantes, científicos e investigadores crear análisis reproducibles y formular una historia dentro de un solo documento.

**Funciones del portátil Jupyter**


1.   Explorador de archivos.
2.   Celdas y sintaxis de Markdown
3.   Núcleos, variables y entorno.
4.   Comando frente a modo de edición y accesos directos.








---
## ¿Qué es Markdown?
---

Markdown es un lenguaje de marcado que utiliza sintaxis de formato de texto sin formato. Esto significa que podemos modificar el formato de nuestro texto con el uso de varios símbolos en nuestro teclado como indicadores.

Para ejemplos de sintaxis y características de Markdown, como tablas, imágenes, etc., los puede encontrar [acá](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)

## Kernels, Variables y Entorno.
---

Un kernel de notebook es un "motor computacional" que ejecuta el código contenido en un documento de Notebook. Hay kernels para varios lenguajes de programación, sin embargo, solo usamos el kernel de **Python** que ejecuta el código de **Python**.


~~~python
print("Esto es una celda de código Python")
~~~

Un kernel es el back-end de nuestro cuaderno que no solo ejecuta el código **Python**, sino que almacena nuestras variables.

```python
### Variable x.
x = 2021
print("x tiene un valor de " + str(x))
```

Es importante tener en cuenta que los Jupyter Notebooks tienen ejecución de celda en línea. Esto significa que una celda en ejecución previa debe completar sus operaciones antes de que se ejecute otra celda. Una celda que aún se está ejecutando se indica con [*] en el lado izquierdo de la celda.

---
## Comando vs. Modo de edición & Accesos directos.
---
Hay un modo de edición y comando para los cuadernos jupyter. El modo es fácilmente identificable por el color del borde izquierdo de la celda.

Azul = Modo de comando.

Verde = Modo de edición.

El modo de comando se puede alternar presionando `esc` en su teclado.

Los comandos se pueden utilizar para ejecutar funciones del cuaderno. Por ejemplo, cambiar el formato de una celda o agregar celdas.
